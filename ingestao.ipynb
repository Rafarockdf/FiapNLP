```python
def connect_to_sql_azure(server, database, username, password):
    connection_string = (
        f"DRIVER={{ODBC Driver 17 for SQL Server}};"
        f"SERVER={server};"
        f"DATABASE={database};"
        f"UID={username};"
        f"PWD={password}"
    )
    return pyodbc.connect(connection_string)
def ingestao(df):
    server = 'enterprise-challenge.database.windows.net'
    database = 'SNIPES'
    username = 'adm'
    password = 'nlp12345!'
    comando_select = 'SELECT Nome,Nome_da_empresa,Unidade_da_TOTVS,Regiao_da_TOTVS,CNPJ,Sentimento,Data_Ligacao,Agilidade_processos_suporte,Atendimento_agentes_suporte,Probabilidade_recomendar_TOTVS, Atendimento_Executivo_Vendas,Custos_valores_produtos,Atendimento_financeiro,Satisfacao_implantacao_software,Experiencia_atualizacao_software,Atendimento_unidade_TOTVS,Recomendacao_empresa, Comentarios FROM avaliacoes;'
    conn = connect_to_sql_azure(server, database, username, password)
    cursor = conn.cursor()
    # Criar a consulta SQL com 18 parâmetros
    comando = """
    INSERT INTO avaliacoes (
        Nome, 
        Nome_da_empresa, 
        Unidade_da_TOTVS, 
        Regiao_da_TOTVS, 
        CNPJ, 
        Sentimento, 
        Data_Ligacao, 
        Agilidade_processos_suporte, 
        Atendimento_agentes_suporte, 
        Probabilidade_recomendar_TOTVS, 
        Atendimento_Executivo_Vendas, 
        Custos_valores_produtos, 
        Atendimento_financeiro, 
        Satisfacao_implantacao_software, 
        Experiencia_atualizacao_software, 
        Atendimento_unidade_TOTVS, 
        Recomendacao_empresa, 
        Comentarios
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?,?)
    """
    for index, row in df.iterrows():
        # Verifica se o registro já existe no banco de dados
        check_query = f"SELECT COUNT(*) FROM avaliacoes WHERE Nome = '{row['Nome']}' AND Nome_da_empresa = '{row['Nome_da_empresa']}' AND Data_Ligacao = '{row['Data_Ligacao']}'"
        cursor.execute(check_query)
        count = cursor.fetchone()[0]

        if count == 0:
            # Se o registro não existe, insere
            valores = [row[coluna] if pd.notna(row[coluna]) else None for coluna in df.columns]
            try:
                cursor.execute(comando, valores)
            except pyodbc.DatabaseError as e:
                print(f"Erro ao executar comando: {e}")
        else:
            print(f"Registro duplicado encontrado para Nome: {row['Nome']}, Empresa: {row['Nome_da_empresa']}")
    conn.commit()
    cursor.close()
    conn.close()
```
